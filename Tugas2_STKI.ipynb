```python
import re
import nltk
import collections
import pprint
from nltk import word_tokenize
from nltk import sent_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

pp = pprint.PrettyPrinter()

text = open("corpus.txt", encoding="ISO-8859-1").read()

factory = StemmerFactory()
stemmer = factory.create_stemmer()
kalimat_stem   = stemmer.stem(text)

factory_s = StopWordRemoverFactory()
stopword = factory_s.create_stop_word_remover()
stop = stopword.remove(kalimat_stem)


token = word_tokenize(stop)

jumlah = collections.Counter(token)
pp.pprint(jumlah)

 
#print(stopwords)
```

```python
import re
import nltk
import collections
import pprint
from nltk import word_tokenize
from nltk import sent_tokenize

pp = pprint.PrettyPrinter()

text = open("corpus.txt", encoding="ISO-8859-1").read()

factory = StemmerFactory()
stemmer = factory.create_stemmer()
kalimat_stem   = stemmer.stem(text)

token = word_tokenize(kalimat_stem)

jumlah = collections.Counter(token).most_common(20)
pp.pprint(jumlah)
```
